```python
def final_score(row):
    score = 0

    # ROUND 1 UPPER
    if row["upper.U5.A"] == "MOI":
        score += 10
    if row["upper.U5.B"] == "BSG":
        score += 10
    if row["upper.U6.A"] == "TMG":
        score += 10
    if row["upper.U6.B"] == "VLD":
        score += 10

    # ROUND 2 UPPER
    if row["upper.U5.A"] == "MOI" and row["upper.U7.A"] == "MOI":
        if row["upper.U5.B"] != "BSG":
            score += 10
        else:
            score += 20

    if row["upper.U6.A"] == "TMG" and row["upper.U7.B"] == "TMG":
        if row["upper.U6.B"] != "VLD":
            score += 10
        else:
            score += 20

    # ROUND 1 Lower
    if row["lower.L1.A"] == "PS" and row["lower.L3.A"] == "PS":
        if row["lower.L1.B"] != "J4F":
            score += 5
        else:
            score += 10

    if row["lower.L1.A"] == "J97" and row["lower.L1.A"] == "CPG":
        if row["lower.L1.B"] != "CPG":
            score += 5
        else:
            score += 10

    return score
```

```python
df["score"] = df.apply(final_score, axis=1)
```

```python
df_result = df[["username", "score", "discordId"]].sort_values(by="score", ascending=False)
```
