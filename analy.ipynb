```python
def final_score(row):
    score = 0

    # ROUND 1 UPPER
    if row["upper.U5.A"] == "MOI":
        score += 5
    if row["upper.U5.B"] == "BSG":
        score += 5
    if row["upper.U6.A"] == "TMG":
        score += 5
    if row["upper.U6.B"] == "VLD":
        score += 5

    # # ROUND 2 UPPER
    # if row["upper.U7.A"] == "MOI":
    #     score += 10

    # if row["upper.U7.B"] == "VLD":
    #     score += 10

    # ROUND 1 Lower
    if row["lower.L1.A"] == "PS" and row["lower.L1.B"] == "J4F":
        if row["lower.L3.A"] == "PS":
            score += 5

    if row["lower.L2.A"] == "J97" and row["lower.L2.B"] == "CPG":
        if row["lower.L4.A"] == "J97":
            score += 5

    return score
```

```python
df["score"] = df.apply(final_score, axis=1)
```

```python
df_result = df[["username", "score"]].sort_values(by="score", ascending=False)
```
