```python
def final_score(row):
    score = 0

    # ROUND 1 UPPER
    if row["upper.U5.A"] == "MOI":
        score += 5
    if row["upper.U5.B"] == "BSG":
        score += 5
    if row["upper.U6.A"] == "TMG":
        score += 5
    if row["upper.U6.B"] == "VLD":
        score += 5

    return score
```

```python
df["score"] = df.apply(final_score, axis=1)
```

```python
df_result = df[["username", "score"]].sort_values(by="score", ascending=False)
```
